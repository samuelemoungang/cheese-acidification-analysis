This notebook analyzes pH traces from two cheese batches:
- **Fromage 1** — USB logger, 1-minute resolution
- **Fromage 2** — Server logger, 1-second resolution

## 1. Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy.signal import savgol_filter

plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 11
```

## 2. Load Data

```python
# --- Fromage 1 (USB, 1-min resolution) ---
df1_raw = pd.read_csv(
    'data/trace pH fromage 1_USB_28042026.csv',
    usecols=['Date', 'Time', 'Ch1_M1'],
)
df1_raw['datetime'] = pd.to_datetime(
    df1_raw['Date'] + ' ' + df1_raw['Time'],
    format='%d/%b/%Y %H:%M:%S'
)
df1_raw['pH'] = pd.to_numeric(df1_raw['Ch1_M1'], errors='coerce')
df1 = df1_raw[['datetime', 'pH']].dropna().set_index('datetime').sort_index()
print(f'Fromage 1: {len(df1)} points  |  {df1.index[0]} → {df1.index[-1]}')
df1.head()
```

```python
# --- Fromage 2 (Server, 1-sec resolution) ---
df2_raw = pd.read_csv(
    'data/trace pH fromage 2_Server_28042026.csv',
    names=['datetime', 'pH'],
    skiprows=1,
)
df2_raw['datetime'] = pd.to_datetime(df2_raw['datetime'])
df2_raw['pH'] = pd.to_numeric(df2_raw['pH'], errors='coerce')
df2 = df2_raw[['datetime', 'pH']].dropna().set_index('datetime').sort_index()
# Keep only physically plausible pH values (3 – 9)
df2 = df2[(df2['pH'] >= 3) & (df2['pH'] <= 9)]
print(f'Fromage 2: {len(df2)} points  |  {df2.index[0]} → {df2.index[-1]}')
df2.head()
```

## 3. Exploratory Overview

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].plot(df1.index, df1['pH'], color='steelblue', linewidth=1.2, label='pH')
axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
axes[0].xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=45)
axes[0].set_title('Fromage 1 — USB logger (1 min)')
axes[0].set_xlabel('Time (28 Apr 2026)')
axes[0].set_ylabel('pH')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Resample Fromage 2 to 1-minute median to reduce noise for display
df2_1min = df2['pH'].resample('1min').median()
axes[1].plot(df2_1min.index, df2_1min.values, color='tomato', linewidth=1.2, label='pH (1-min median)')
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
axes[1].xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45)
axes[1].set_title('Fromage 2 — Server logger (1 sec → 1 min median)')
axes[1].set_xlabel('Time (28 Apr 2026)')
axes[1].set_ylabel('pH')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.suptitle('Acidification Profiles — 28 April 2026', fontweight='bold')
plt.tight_layout()
plt.savefig('acidification_overview.png', bbox_inches='tight')
plt.show()
```

## 4. Descriptive Statistics

```python
stats = pd.DataFrame({
    'Fromage 1 (USB)': df1['pH'].describe(),
    'Fromage 2 (Server)': df2['pH'].describe(),
}).T
stats
```

## 5. Acidification Rate (dpH/dt)

```python
# Fromage 1 — rate between consecutive 1-min samples
dt1_min = df1.index.to_series().diff().dt.total_seconds() / 60
df1['dpH_dt'] = df1['pH'].diff() / dt1_min

# Fromage 2 — resample to 1-min median, then compute rate
df2_rate = df2['pH'].resample('1min').median().to_frame(name='pH')
dt2_min = df2_rate.index.to_series().diff().dt.total_seconds() / 60
df2_rate['dpH_dt'] = df2_rate['pH'].diff() / dt2_min

fig, axes = plt.subplots(1, 2, figsize=(16, 4))

axes[0].plot(df1.index, df1['dpH_dt'], color='steelblue', linewidth=0.8, alpha=0.7)
axes[0].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
axes[0].xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=45)
axes[0].set_title('Fromage 1 — Acidification rate')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('dpH/dt (pH units / min)')
axes[0].grid(alpha=0.3)

axes[1].plot(df2_rate.index, df2_rate['dpH_dt'], color='tomato', linewidth=0.8, alpha=0.7)
axes[1].axhline(0, color='black', linewidth=0.5, linestyle='--')
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
axes[1].xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45)
axes[1].set_title('Fromage 2 — Acidification rate')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('dpH/dt (pH units / min)')
axes[1].grid(alpha=0.3)

plt.suptitle('Acidification Rate (dpH/dt)', fontweight='bold')
plt.tight_layout()
plt.savefig('acidification_rate.png', bbox_inches='tight')
plt.show()
```

## 6. Comparative Plot (aligned from start)

```python
# Align both series so t=0 is the first measurement
t1 = (df1.index - df1.index[0]).total_seconds() / 3600
t2_median = df2_1min
t2 = (t2_median.index - t2_median.index[0]).total_seconds() / 3600

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(t1, df1['pH'].values, color='steelblue', linewidth=1.4, label='Fromage 1 (USB)')
ax.plot(t2, t2_median.values, color='tomato', linewidth=1.4, label='Fromage 2 (Server, 1-min median)')
ax.set_xlabel('Time since start (hours)')
ax.set_ylabel('pH')
ax.set_title('Acidification Profiles — Time-aligned Comparison', fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('acidification_comparison.png', bbox_inches='tight')
plt.show()
```

## 7. Rolling Statistics (Fromage 2)

```python
# 10-minute rolling window on Fromage 2 (server data has sub-minute noise)
df2_roll = df2['pH'].resample('1min').median()
roll_mean = df2_roll.rolling(10, center=True).mean()
roll_std  = df2_roll.rolling(10, center=True).std()

fig, ax = plt.subplots(figsize=(14, 5))
ax.fill_between(roll_mean.index,
                roll_mean - roll_std,
                roll_mean + roll_std,
                alpha=0.25, color='tomato', label='±1 std (10 min window)')
ax.plot(roll_mean.index, roll_mean, color='tomato', linewidth=1.4, label='Rolling mean')
ax.plot(df2_roll.index, df2_roll.values, color='grey', linewidth=0.5, alpha=0.5, label='1-min median')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax.xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
ax.set_title('Fromage 2 — Rolling mean & variability', fontweight='bold')
ax.set_xlabel('Time (28 Apr 2026)')
ax.set_ylabel('pH')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('fromage2_rolling.png', bbox_inches='tight')
plt.show()
```

## 8. Key Milestones

Identify when each batch crosses critical pH thresholds (e.g. pH 5.0 — typical curd set point).

```python
thresholds = [5.2, 5.0, 4.8, 4.6]

def first_crossing(series, threshold):
    """Return the first timestamp where pH <= threshold."""
    below = series[series <= threshold]
    if below.empty:
        return None
    return below.index[0]

rows = []
for thr in thresholds:
    t_f1 = first_crossing(df1['pH'], thr)
    t_f2 = first_crossing(df2_1min, thr)
    rows.append({
        'pH threshold': thr,
        'Fromage 1 reached at': t_f1.strftime('%H:%M') if t_f1 else 'not reached',
        'Fromage 2 reached at': t_f2.strftime('%H:%M') if t_f2 else 'not reached',
    })

milestones = pd.DataFrame(rows)
milestones
```
